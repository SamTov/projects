```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def rotate_vector(theta: float, x: np.ndarray):
    """
    Rotate the vector.
    
    Parameters
    ----------
    theta : float
            Angle by which you want to rotate.
    x : np.ndarray
            Vector to rotate.
    """
    theta = np.deg2rad(theta)
    matrix = np.array(
        [
            [np.cos(theta), np.sin(theta)],
            [-np.sin(theta), np.cos(theta)],
        ]
    )
    
    return matrix @ x
```

```python
def compute_angle(current_time, previous_time):
    """
    Compute the angle between two vectors
    """
    angle = np.arctan2(
        previous_time[0] * current_time[1] - previous_time[1] * current_time[0],
        previous_time[0] * current_time[0] + previous_time[1] * current_time[1] 
    )
    
    return np.rad2deg(angle)
```

```python
seed_vector = np.array([0., 1.])
```

```python
trajectory = [seed_vector]
```

```python
vector = seed_vector
for _ in range(10):
    vector = rotate_vector(5., vector)
    trajectory.append(
        vector
    )
```

```python
for item in trajectory:
    plt.plot(item[0], item[1], "o")
    
plt.xlim(-1., 1.)
plt.ylim(-1., 1.)
plt.show()
```
