```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import h5py as hf
from matplotlib.colors import LogNorm
```

# Single Well

```python
with hf.File("single-well/trajectory.hdf5", "r") as db:
    s_positions = db["colloids"]["Unwrapped_Positions"][:]
```

```python
sns.displot(
    x=s_positions[:200, :, 0].flatten(),
    y=s_positions[:200, :, 1].flatten(),
    norm=LogNorm(),
    vmin=None,
    vmax=None,
    bins=500,
    cmap="coolwarm",
    kind="hist"
)
plt.savefig("single-well.png")
plt.show()
```

# Double Well

```python
with hf.File("double-well/trajectory.hdf5", "r") as db:
    d_positions = db["colloids"]["Unwrapped_Positions"][:]
```

```python
sns.displot(
    x=d_positions[:, :, 0].flatten(),
    y=d_positions[:, :, 1].flatten(),
    norm=LogNorm(),
    vmin=None,
    vmax=None,
    bins=500,
    cmap="coolwarm",
    kind="hist"
)
plt.savefig("double-well.png")
plt.show()
```
