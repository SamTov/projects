```python
import znvis as vis
import h5py as hf
import numpy as np
```

```python
with hf.File("double_well/trajectory.hdf5", "r") as db:
    data = db["colloids"]["Unwrapped_Positions"][:]
```

```python
colloid_mesh = vis.Sphere(radius=2.14, colour=np.array([30, 144, 255]) / 255, resolution=2)
colloid_particle = vis.Particle(name="Colloid", mesh=colloid_mesh, position=data)
```

```python
visualizer = vis.Visualizer(particles=[colloid_particle], frame_rate=80)
visualizer.run_visualization()
```
