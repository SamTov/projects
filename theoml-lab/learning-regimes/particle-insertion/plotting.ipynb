```python
import numpy as np
from scipy.optimize import curve_fit
from scipy.ndimage import uniform_filter1d

import matplotlib.pyplot as plt
import seaborn as sns

from dataclasses import dataclass
```

```python
@dataclass
class Measurement:
    width: int
    depth: int
    loss: np.ndarray
    trace: np.ndarray
    entropy: np.ndarray
    representations: np.ndarray
    loss_derivatives: np.ndarray
    class_entropy: dict
    n_particles: int
```

```python
results = np.load("large-experiment-analysis.npy", allow_pickle=True)
```

# Loss

```python
for item in results:
    plt.plot(item.loss, label=item.n_particles)
    

# Axis parameters
plt.yscale("log")

plt.xlabel("Epoch")
plt.ylabel("Train Loss")

# Show parameters    
plt.legend()
plt.show()
```

## Trace

```python
for item in results:
    plt.plot(item.trace / (2 * item.n_particles), label=item.n_particles)
    

# Axis parameters
plt.yscale("log")
plt.xscale("log")

plt.xlabel("Epoch")
plt.ylabel("Trace")

# Show parameters    
plt.legend()
plt.show()
```

## Entropy

```python
for item in results:
    plt.plot(item.entropy, label=item.n_particles)
#     plt.hlines(np.log(2 * item.n_particles), 0, 10000)
    

# Axis parameters
# plt.yscale("log")
# plt.xscale("log")

plt.xlabel("Epoch")
plt.ylabel("Entropy")

# Show parameters    
plt.legend()
plt.show()
```

## Loss Derivative

```python
for item in results:
    plt.plot(item.loss_derivatives.mean(axis=1), label=item.n_particles)
    

# Axis parameters
# plt.yscale("log")
# plt.xscale("log")

plt.xlabel("Epoch")
plt.ylabel("Loss Derivative")
plt.xlim(0, 100)
# Show parameters    
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax1 = ax.twinx()

ax.plot(
    np.mean(results[-1].loss_derivatives, axis=1).sum(axis=-1), 
    label="LD",
#     alpha=0.3
)
# ax.plot(
#     uniform_filter1d(
#         results[-1].loss_derivatives.sum(axis=-1).sum(axis=-1),
#         size=100
#     ),
#     label="Av-LD"
# )

ax1.plot((results[-1].trace / (2 * item.n_particles)), label="Trace", c="green")  

# Axis parameters
# plt.yscale("log")
# plt.xscale("log")

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss Derivative")

ax1.set_ylabel("Trace")
ax1.set_yscale("log")

ax.set_xlim(0, 500)
# Show parameters    
fig.legend(loc=(0.3, 0.7))

plt.savefig("ld_and_trace_short_mean.pdf")
plt.show()
```
