```python
import numpy as np
import matplotlib.pyplot as plt
import h5py as hf
```

```python
def load_data(architecture: str, optimizer: str, data_split: str):
    loss = []
    accuracy = []
    for i in range(9):
        with hf.File(f"{architecture}/{optimizer}/{data_split}_recorder_{i}.h5") as db:
            loss.append(db["loss"][:5000])
            accuracy.append(db["accuracy"][:5000])
            plt.plot(db["accuracy"][:])
            plt.title(max(db["accuracy"][:]))
            plt.show()
            
    mean_loss = np.mean(loss, axis=0)
    err_loss = np.std(loss, axis=0) / len(loss)
    
    mean_accuracy = np.mean(accuracy, axis=0)
    err_accuracy = np.std(accuracy, axis=0) / len(accuracy)
    
    
    return mean_loss, err_loss, mean_accuracy, err_accuracy
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 15))


epochs = np.linspace(1, 5000, 5000)

# Accuracy
ax[0].plot(epochs, adam_data[2], label="ADAM")
ax[0].fill_between(epochs, adam_data[2]-adam_data[3], adam_data[2]+adam_data[3])

ax[0].plot(epochs, sgd_data[2], label="SGD")
ax[0].fill_between(epochs, sgd_data[2]-sgd_data[3], sgd_data[2]+sgd_data[3])

ax[0].plot(epochs, traceopt[2], label="TraceOpt")
ax[0].fill_between(epochs, traceopt[2]-traceopt[3], traceopt[2] + traceopt[3])

# ax[0].set_xscale("log")

# Loss

ax[1].plot(epochs, adam_data[0], label="ADAM")
ax[1].fill_between(epochs, adam_data[0]-adam_data[1], adam_data[0]+adam_data[1])

ax[1].plot(epochs, sgd_data[0], label="SGD")
ax[1].fill_between(epochs, sgd_data[0]-sgd_data[1], sgd_data[0]+sgd_data[1])

ax[1].plot(epochs, traceopt[0], label="TraceOpt")
ax[1].fill_between(epochs, traceopt[0]-traceopt[1], traceopt[0] + traceopt[1])
# ax[1].set_xlim(0, 4)

fig.legend()
plt.show()
```
