```python
# Linalg
import numpy as np 

# Plotting
import matplotlib.pyplot as plt

# Data
import h5py as hf
```

```python
with hf.File("train_resnet18_55452.h5", "r") as db:
    train_loss = db["loss"][:]
    train_acc = db["accuracy"][:]
    entropy = db["entropy"][:]
    trace = db["trace"][:]

with hf.File("test_resnet18_55452.h5", "r") as db:
    test_loss = db["loss"][:]
    test_acc = db["accuracy"][:]
    test_entropy = db["entropy"][:]
    test_trace = db["trace"][:]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))


ax[0].plot(train_acc, label="Train")
ax[0].plot(test_acc, label="Test")

ax[0].set_ylabel("Accuracy")
ax[0].legend()
ax[0].set_xlabel("Epochs")

ax[1].plot(entropy, label="Train")
ax[1].plot(test_entropy, label="Test")

ax[1].legend()
ax[1].set_ylabel("Entropy")
ax[1].set_xlabel("Epochs")

ax[2].plot(trace, label="Train")
ax[2].plot(test_trace, label="Test")

# Set y axis to log scale
ax[2].set_yscale("log")
ax[2].set_ylabel("Trace")
ax[2].set_xlabel("Epochs")

ax[2].legend()

plt.savefig("resenet18.pdf")
plt.show()
```
