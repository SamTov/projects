```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def running_average(data, window_size=10):
    return pd.Series(data).rolling(window=window_size, min_periods=1).mean().to_numpy()

fig, axes = plt.subplots(5, 3, figsize=(8.0, 8.0))
colors = ['#E6AF2E', '#30C5FF', '#09BC8A']

# ResNet
ax = axes[0]
entropy = np.load("resnet-entropy.npy", allow_pickle=True)
trace = np.load("resnet-trace.npy", allow_pickle=True)
loss = np.load("resnet-loss.npy", allow_pickle=True)

for i, (data, ylabel, title) in enumerate(zip([entropy, trace, loss], ["Entropy", "Trace", "Loss"], ["a) ResNet, CIFAR10", "", ""])):
    data = np.array(data)
    x = np.linspace(0, 180, len(data))
    y = data[:, 0] if data.ndim > 1 else data
    ax[i].plot(x, y, color=colors[i], alpha=0.3)
    ax[i].plot(x, running_average(y), color=colors[i])
    ax[i].set_ylabel(ylabel)
    ax[i].set_title(title, loc='left')
ax[1].set_yscale("log")

# Transformer
ax = axes[1]
entropy = np.load("nlp-entropy.npy", allow_pickle=True)
trace = np.load("nlp-trace.npy", allow_pickle=True)
loss_data = np.load("nlp-loss.npy", allow_pickle=True)
loss = [np.mean([item[0]["loss"] / item[0]["denominator"], item[1]["loss"] / item[1]["denominator"]]) for item in loss_data]

for i, (data, ylabel, title) in enumerate(zip([entropy, trace, loss], ["Entropy", "Trace", "Loss"], ["b) Transformer, NLP Translation", "", ""])):
    data = np.array(data)
    x = np.linspace(0, 75000, len(data))
    y = data[:, 0] if data.ndim > 1 else data
    ax[i].plot(x, y, color=colors[i], alpha=0.3)
    ax[i].plot(x, running_average(y), color=colors[i])
    ax[i].set_ylabel(ylabel)
    ax[i].set_title(title, loc='left')
ax[1].set_yscale("log")

# RL
ax = axes[2]
actor_entropy = np.load("actor_entropy.npy")
actor_trace = np.load("actor_traces.npy")
critic_entropy = np.load("critic_entropy.npy")
critic_trace = np.load("critic_traces.npy")
rewards = np.load("rewards.npy")

x = np.linspace(0, 39062, len(actor_entropy))
ax[0].plot(x, actor_entropy[:, 0], color=colors[0], alpha=0.3, label='Actor')
ax[0].plot(x, running_average(actor_entropy[:, 0]), color=colors[0])
ax[0].plot(x, critic_entropy[:, 0], color=colors[1], alpha=0.3, linestyle='--', label='Critic')
ax[0].plot(x, running_average(critic_entropy[:, 0]), color=colors[1])

ax[1].plot(x, actor_trace[:, 0], color=colors[0], alpha=0.3)
ax[1].plot(x, running_average(actor_trace[:, 0]), color=colors[0])
ax[1].plot(x, critic_trace[:, 0], color=colors[1], alpha=0.3, linestyle='--')
ax[1].plot(x, running_average(critic_trace[:, 0]), color=colors[1])

ax[2].plot(x, rewards, color=colors[2], alpha=0.3)
ax[2].plot(x, running_average(rewards), color=colors[2])

for i, ylabel in enumerate(["Entropy", "Trace", "Reward"]):
    ax[i].set_ylabel(ylabel)
ax[0].legend()
ax[1].set_yscale("log")
ax[1].set_xscale("log")
ax[0].set_xscale("log")
ax[2].set_xscale("log")
ax[0].set_title("c) RL, Pong", loc='left')

# GNN
ax = axes[3]
entropy = np.load("gnn-entropy.npy", allow_pickle=True)
trace = np.load("gnn-trace.npy", allow_pickle=True)
loss = np.load("gnn-loss.npy", allow_pickle=True)
loss = [item["loss"] for item in loss]

for i, (data, ylabel, title) in enumerate(zip([entropy, trace, loss], ["Entropy", "Trace", "Loss"], ["d) GNN, Molecule prediction", "", ""])):
    data = np.array(data)
    x = np.linspace(0, 100000, len(data))
    y = data[:, 0] if data.ndim > 1 else data
    ax[i].plot(x, y, color=colors[i], alpha=0.3)
    ax[i].plot(x, running_average(y), color=colors[i])
    ax[i].set_ylabel(ylabel)
    ax[i].set_title(title, loc='left')
ax[1].set_yscale("log")

# VAE
ax = axes[4]
entropy = np.load("vae-entropy.npy", allow_pickle=True)
trace = np.load("vae-trace.npy", allow_pickle=True)
loss = np.load("vae-loss.npy", allow_pickle=True)

for i, (data, ylabel, title) in enumerate(zip([entropy, trace, loss], ["Entropy", "Trace", "Loss"], ["e) VAE, MNIST", "", ""])):
    data = np.array(data)
    x = np.linspace(0, 200, len(data))
    y = data[:, 0] if data.ndim > 1 else data
    ax[i].plot(x, y, color=colors[i], alpha=0.3)
    ax[i].plot(x, running_average(y), color=colors[i])
    ax[i].set_ylabel(ylabel)
    ax[i].set_title(title, loc='left')
    ax[i].set_xlabel("Epoch")
ax[1].set_yscale("log")

plt.tight_layout()
plt.savefig("novely-architectures.pdf")
plt.show()
```
