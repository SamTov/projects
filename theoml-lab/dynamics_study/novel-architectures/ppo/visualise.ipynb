```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
actor_entropy = np.load("actor_entropy.npy")
actor_trace = np.load("actor_traces.npy")

critic_entropy = np.load("critic_entropy.npy")
critic_trace = np.load("critic_traces.npy")

rewards = np.load("rewards.npy")
```

```python
fig, axe = plt.subplots(1, 2, figsize=(12, 5))


ax = axe[0]
ax2 = ax.twinx()

ax.plot(np.linspace(0, rewards.shape[0] * 100, rewards.shape[0]), rewards, '.')
ax2.errorbar(np.linspace(0, rewards.shape[0] * 100, rewards.shape[0]), actor_entropy[:, 0], color="orange", marker="o", label="Actor", markersize=1)
ax2.errorbar(np.linspace(0, rewards.shape[0] * 100, rewards.shape[0]), critic_entropy[:, 0], color="black", marker="o", label="Critic", markersize=1)
ax2.legend()
ax.set_ylabel("Rewards")
ax2.set_ylabel("Entropy")
ax.set_xlabel("Episode")

ax = axe[1]
ax2 = ax.twinx()

ax.plot(np.linspace(0, rewards.shape[0] * 100, rewards.shape[0]), rewards, '.')
ax2.errorbar(np.linspace(0, rewards.shape[0] * 100, rewards.shape[0]), actor_trace[:, 0], color="orange", marker="o", markersize=1)
ax2.errorbar(np.linspace(0, rewards.shape[0] * 100, rewards.shape[0]), critic_trace[:, 0], color="black", marker="o", markersize=1)
ax2.set_yscale("log")
ax2.set_ylabel("Trace")
ax.set_xlabel("Episode")


plt.savefig("ppo.pdf")
plt.show()
```
