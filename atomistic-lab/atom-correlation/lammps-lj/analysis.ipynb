```python
import mdsuite as mds
from mdsuite.utils import Units

import glob
import matplotlib.pyplot as plt

import h5py as hf
```

```python
units = Units(
        time=1,
        length=1,
        energy=1,
        NkTV2p=1,
        boltzmann=1,
        temperature=1,
        pressure=1,
    )
```

```python
project = mds.Project("LJ")
```

```python
for item in sim_directories:
    project.add_experiment(
        name=item,
        timestep=0.005,
        temperature=float(item.split("_")[1]),
        units=units,
        simulation_data=f"{item}/dumpy.lammpstraj"
    )
```

```python
project.run.EinsteinDiffusionCoefficients(
    data_range=1000, correlation_time=50
)
```

```python
density = []
pressure = []
for item in project.experiments:
    pressure.append(float(item.split("_")[1]))
    volume = project.experiments[item].volume
    density.append(4000 / volume)
```

```python
plt.plot(pressure, density, '.')
```
