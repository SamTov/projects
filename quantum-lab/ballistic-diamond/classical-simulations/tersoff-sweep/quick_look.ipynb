```python
from glob import glob

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
files = glob("/work/stovey/ballistic-diamond/tersoff-sweep/energy-35/temperature-300/*")
```

```python
files[0]
```

```python
with open(f"{files[1]}/collision-ion.lammpstraj") as f:
    lines = f.readlines()
    start = float(lines[9].split()[4])
    end = float(lines[-1].split()[4])
```

```python
depths = []
errors = 0
for it in files:
    try:
        with open(f"{it}/collision-ion.lammpstraj") as f:
            lines = f.readlines()
            start = float(lines[9].split()[4])
            end = float(lines[-1].split()[4])
            depths.append(start - end)
    except Exception as e:
        errors += 1
```

```python
sns.histplot(np.array(depths) / 10, kde=True)
plt.xlabel("Depth (nm)")
```
