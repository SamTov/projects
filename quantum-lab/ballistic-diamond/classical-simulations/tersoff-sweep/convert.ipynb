```python
import numpy as np
```

```python
# 20 keV 40 keV 80 keV 170 keV
energies = np.array([20, 40, 60, 80, 100, 170])
sn_mass = 118.71
avogadro = 6.02214076e23
joule = 1.602176634e-19
angstorm = 1e-10
picosecond = 1e-12
```

```python
def compute_velocity(energy, mass):
    kg_mass = (mass / avogadro) / 1000
    ev_energy = (energy * 1000) * joule

    mps = np.sqrt(2 * ev_energy / kg_mass)
    return mps * picosecond / angstorm
```

```python
velocities = compute_velocity(energies, sn_mass)
```
