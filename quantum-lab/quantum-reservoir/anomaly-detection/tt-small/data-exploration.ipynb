```python
# Linalg libraries
import numpy as np

# Plotting libraries
import matplotlib.pyplot as plt
```

```python
background = np.genfromtxt(
    "./data/background.csv", delimiter=",", skip_header=1
)
signal = np.genfromtxt(
    "./data/signal.csv", delimiter=",", skip_header=1
)
```

```python
plt.plot(background[:,1], background[:, 2], '.', label="Background")
plt.plot(signal[:,1], signal[:, 2], '.', label="Signal")
plt.legend()
plt.show()
```
