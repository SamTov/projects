```python
def init(symbol, state: blankly.StrategyState):
    # Download price data to give context to the algo
    state.variables['history'] = state.interface.history(
        symbol, to=150, return_as='deque'
    )['close']
    state.variables["count"] = 0
    state.variables["available_currency"] = 10000
```

```python
# Authenticate Oanda strategy
exchange = blankly.Oanda()

# Use our strategy helper on Oanda
strategy = blankly.Strategy(exchange)

strategy.add_price_event(price_event, symbol='EUR-AUD', resolution='1h', init=init)
strategy.add_price_event(price_event, symbol='AUD-USD', resolution='1h', init=init)
strategy.add_price_event(price_event, symbol='EUR-USD', resolution='1h', init=init)


strategy.backtest(to='1y', initial_values={'EUR': 10000})
```
