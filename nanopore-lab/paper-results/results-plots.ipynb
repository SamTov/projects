```python
import numpy as np
import torch

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
resnet18 = np.load("cm-resnet-18.npy")
resnet101 = np.load("cm-resnext-101.npy")
transformer = torch.load("cm-transformer.pt").numpy()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(8.0, 5.0))

## Confusion Matrices ##
# Resnet18
ax = axes[0, 0]
sns.heatmap(
    resnet18 / resnet18.sum(axis=1), 
    cmap="Blues", 
    yticklabels=[], 
    xticklabels=[], 
    ax=ax, 
    vmin=0, 
    vmax=1.0, 
    cbar=False
)
ax.set_title("Resnet 18")

# Resnext101
ax = axes[0, 1]
sns.heatmap(
    resnet101 / resnet101.sum(axis=1), 
    cmap="Blues", 
    yticklabels=[], 
    xticklabels=[], 
    ax=ax, 
    vmin=0, 
    vmax=1.0, 
    cbar=False
)
ax.set_title("Resnext 101")

# Transformer
ax = axes[0, 2]
sns.heatmap(
    transformer / transformer.sum(axis=1), 
    cmap="Blues", 
    yticklabels=[], 
    xticklabels=[], 
    ax=ax, 
    vmin=0, 
    vmax=1.0, 
    cbar=False
)
ax.set_title("Vision Transformer")

# Colourbar
cbar_ax = fig.add_axes([0.92, 0.53, 0.01, 0.35])  # [left, bottom, width, height]
plt.colorbar(ax.collections[0], cax=cbar_ax)

## Histograms ##
bins = 15

# Resnet18
ax = axes[1, 0]
sns.histplot(np.diag(resnet18 / resnet18.sum(axis=1)), bins=bins, kde=True, ax=ax, stat="percent")
ax.set_xlim(0.0, 1.0)
ax.set_ylabel("% Occurrence")

# Resnext101
ax = axes[1, 1]
pl = sns.histplot(np.diag(resnet101 / resnet101.sum(axis=1)), bins=bins, kde=True, ax=ax)
pl.set(ylabel=None)
ax.set_xlim(0.0, 1.0)

# Transformer
ax = axes[1, 2]
pl = sns.histplot(np.diag(transformer / transformer.sum(axis=1)), bins=bins, kde=True, ax=ax)
ax.set_xlim(0.0, 1.0)
pl.set(ylabel=None)

plt.savefig("classification-results.pgf")
plt.show()
```
