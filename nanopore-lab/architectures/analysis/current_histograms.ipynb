```python
# Linear algebra
import numpy as np

# Plotting
import seaborn as sns
import matplotlib.pyplot as plt

# File management
import pickle
import glob
```

```python
root_dir = "/data/jhossbach/Peptide_events"
```

```python
files = glob.glob(f"{root_dir}/*.pk")
```

```python
for file in files:
    peptide = file.split("/")[-1].split(".")[0]
    if peptide in ["L1AS5", "L2AS5", "L3AS5", "L5AS5", "L6AS5"]:
        with open(file, "br") as f:
            data = pickle.load(f)
            sns.kdeplot(np.concatenate(data), label=peptide, fill=True)


plt.xlim(0.0, 0.85)
plt.xlabel("Normalized Current Amplitude")
plt.ylabel("Density")
plt.legend(loc=[0.1, 0.5])
plt.savefig("current_histograms.png", dpi=600)
plt.show()
```
